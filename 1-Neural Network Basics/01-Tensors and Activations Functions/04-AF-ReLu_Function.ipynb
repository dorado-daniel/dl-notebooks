# ReLu Function

La ReLU (Rectified Linear Unit) es la funci√≥n de activaci√≥n m√°s usada en redes neuronales modernas.

Es decir:

Si la salida del perceptr√≥n 
ùë¶<0  devuelve 0 (la neurona ‚Äúno se activa‚Äù).

Si ùë¶‚â•0 devuelve y (la neurona ‚Äúse activa‚Äù proporcionalmente).

```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

# === Datos de entrada: 3 muestras, 2 caracter√≠sticas ===
X = torch.tensor([
    [2.0, 1.0],
    [5.0, 1.0],
    [1.0, 3.0]
], device=device)

# === Pesos y bias ===
W = torch.tensor([[4.0], [-3.0]], device=device)
b = torch.tensor([-5.0], device=device)

# === C√°lculo del perceptr√≥n ===
# y es la salida lineal antes de aplicar activaci√≥n
y = X @ W + b

print("Resultado del c√°lculo del perceptr√≥n (y):\n", y)
```

```python
# === Funci√≥n de activaci√≥n ReLU ===
f = torch.nn.ReLU()

# Aplicamos ReLU sobre la salida del perceptr√≥n
y_relu = f(y)

print("Salida lineal (pre-activaci√≥n):\n", y)
print("\nSalida tras ReLU:\n", y_relu)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))

# === Dibujamos la curva ReLU ===
x_plot = torch.linspace(-15, 15, 400)
y_plot = torch.where(x_plot >= 0, x_plot, torch.tensor(0.0))
plt.plot(x_plot.cpu(), y_plot.cpu(), label="ReLU Function", color="blue")

# === A√±adimos las salidas reales del perceptr√≥n ===
plt.scatter(y.cpu(), y_relu.cpu(), color="red", s=80, label="Salidas del perceptr√≥n")

plt.title("ReLU aplicada a la salida lineal del perceptr√≥n")
plt.xlabel("Salida lineal (y)")
plt.ylabel("Salida ReLU f(y)")
plt.grid(True)
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# === Derivada de ReLU ===
x_plot = torch.linspace(-10, 10, 400)
dy_dx = torch.where(x_plot >= 0, torch.tensor(1.0), torch.tensor(0.0))

plt.figure(figsize=(7, 5))
plt.plot(x_plot.cpu(), torch.where(x_plot >= 0, x_plot, torch.tensor(0.0)).cpu(), label="f(x) = ReLU", color="blue")
plt.plot(x_plot.cpu(), dy_dx.cpu(), "--", label="f'(x) = Derivada", color="orange")

plt.title("ReLU y su derivada")
plt.xlabel("x (resultado del c√°lculo del perceptr√≥n)")
plt.ylabel("Valor")
plt.grid(True)
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.legend()
plt.show()
```
