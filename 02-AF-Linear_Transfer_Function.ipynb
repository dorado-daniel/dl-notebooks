También se la llama Identity Function, es una no-funcion de activación, ya que no introduce no linearidad, es el mismo ejemplo del notebook anterior.

```python
import torch

# === Datos de entrada ===
x_houses = torch.tensor([
    [90.0, 3.0],
    [120.0, 4.0],
    [60.0, 2.0]
], device="cuda")

# === Pesos y bias ===
w = torch.tensor([
    [0.8],
    [0.3]
], device="cuda")

b = torch.tensor([10.0], device="cuda")

# === Operación lineal ===
y = x_houses @ w + b
print("Salida lineal (y = x @ w + b):")
print(y)
```

```python
 #Vamos a graficar la función identidad junto con las salidas del perceptrón
import matplotlib.pyplot as plt

# === Función identidad ===
x_func = torch.linspace(0, 120, 100)  # mismo rango que los datos reales
y_func = x_func  # f(x) = x


y_houses = x_houses @ w + b

# === Gráfica ===
plt.figure(figsize=(7, 5))
plt.plot(x_func, y_func, label="f(x) = x", color="blue")
plt.scatter(y_houses.cpu(), y_houses.cpu(), color="red", s=60, label="Salidas del perceptrón")
plt.title("Identity (Linear) Transfer Function con salidas del perceptrón")
plt.xlabel("Entrada (x)")
plt.ylabel("Salida (f(x))")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend()
plt.show()
```

```python


# Dominio de la función
x = torch.linspace(-10, 10, 100, requires_grad=True)

# Función identidad: f(x) = x
y = x

# Derivada analítica: df/dx = 1 en todo punto
dy_dx = torch.ones_like(x)

# Visualización
plt.figure(figsize=(6, 4))
plt.plot(x.detach(), y.detach(), label="f(x) = x", color="blue")
plt.plot(x.detach(), dy_dx.detach(), label="df/dx = 1", color="orange", linestyle="--")
plt.title("Identity Function y su Derivada")
plt.xlabel("x")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.axhline(0, color="black", linewidth=0.8)
plt.show()


# La derivada es plana porque la derivada f'(x) = 1 para todo x:
# La derivada mide la tasa de cambio de la función.
# En la función identidad, la tasa de cambio es constante (1) en todo el dominio

```
